```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import datetime
```

```python
eth_df = pd.read_json('ethereum_prices.json')
eth_df
```

```python
eth_df.dtypes
```

```python
eth_df = eth_df.iloc[1:,:]
eth_df.reset_index(inplace=True,drop=True)
eth_df
```

```python
twit_df = pd.read_csv('sentimentExtractedTweets.txt',header=None)
```

```python
twit_df.head()
```

```python
twit_df.dtypes
```

```python
datetime_simp = []
for i, row in twit_df.iterrows():
    hour = int(twit_df.iloc[i,1][:2])
    if hour < 4:
        datetime_simp.append('00:00:00')
    elif hour < 8:
        datetime_simp.append('04:00:00')
    elif hour < 12:
        datetime_simp.append('08:00:00')
    elif hour < 16:
        datetime_simp.append('12:00:00')
    elif hour < 20:
        datetime_simp.append('16:00:00')
    elif hour < 24:
        datetime_simp.append('20:00:00')
```

```python
twit_df.iloc[:,1] = datetime_simp
twit_df
```

```python
twit_df['datetime'] = twit_df.iloc[:,0] + ' ' + twit_df.iloc[:,1]
twit_df.head()
```

```python
twit_df['datetime'] = pd.to_datetime(twit_df['datetime'])
```

```python
twit_df.sort_values('datetime',axis=0,ascending=True)
```

```python
twit_df.drop([0,1],axis=1, inplace = True)
```

```python
twit_df.columns = ['score', 'datetime']
twit_df
```

```python
mean = twit_df.groupby('datetime').mean()
```

```python
std = twit_df.groupby('datetime').std()
```

```python
eth_df = eth_df.set_index('date').join(mean,how='left',rsuffix='_mean').join(std, how='left', lsuffix='_std')
```

```python
eth_df
```

```python
eth_df.drop('weightedAverage',axis=1,inplace=True)
eth_df.head()
```

```python
eth_df.fillna(eth_df.max(),inplace=True)
```

```python
eth_np = eth_df.as_matrix()
```

```python
from sklearn.preprocessing import scale
```

```python
X = eth_np[:-1,:]
y = eth_np[1:,0]
```

```python
X = scale(X, axis=0, with_mean=True, with_std=True, copy=True)
```

```python
train = eth_np[:4469,:]
test = eth_np[4469:, :]
```

```python
eth_np.shape
```

```python
X_train = X[:4469]
y_train = y[:4469]
X_test = X[4469:]
y_test = y[4469:]
```

```python
X_train.shape
```

```python
y_test
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
from keras.layers.core import Dense, Dropout, Flatten, Activation
from keras.layers.recurrent import LSTM
from keras.models import Sequential
```

```python
from keras import backend as K
from keras.utils.generic_utils import get_custom_objects

def custom_activation(x):
    return abs(x)

get_custom_objects().update({'custom_activation': Activation(custom_activation)})

```

```python
model = Sequential()

model.add(LSTM(50, input_shape = (X_train.shape[1], X_train.shape[2]),
              return_sequences = True))

model.add(LSTM(100, return_sequences = True))
model.add(LSTM(100, return_sequences = True))
model.add(Dropout(0.2))

model.add(Flatten())

model.add(Dense(50, activation='linear'))
model.add(Dense(25, activation='linear'))
model.add(Dense(10, activation='linear'))
model.add(Dense(5, activation='linear'))
model.add(Dense(1, activation='linear'))
model.add(Activation(custom_activation))
```

```python
model.compile(optimizer='Adam',loss='mae')
```

```python
X_train
```

```python
hist = model.fit(
    X_train,
    y_train,
    validation_split = 0.1)
```

```python
model.evaluate(X_test, y_test)
```

```python
from matplotlib import pyplot as plt

plt.plot(hist.history['loss'], label='train')
plt.plot(hist.history['val_loss'], label='val')
plt.legend()
plt.show()
```

```python
predict = model.predict(X_test)
```

```python
plt.plot(predict, label='prediction')
plt.plot(y_test, label='actual')
plt.legend()
plt.savefig('prediction_graph')
plt.show()
```
